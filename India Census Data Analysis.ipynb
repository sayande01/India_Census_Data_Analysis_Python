# 1. State wise Total population

```python
df1 = df.groupby('State_name')['Population'].sum().sort_values(ascending = False).head(10)
df1
```

```python
plt.xticks(rotation=90)
ax = sns.barplot(x = df1.index, y = df1.values, palette = 'dark')
ax.ticklabel_format(style='plain', axis='y')
```

# 2. Male Female Distribution

```python
df2 = df[['Male','Female']].sum()
df2
```

```python
df2.plot(y = df2.values,kind = 'pie',autopct= '%1.1f%%')
plt.show()
```

```python
df10 = df.groupby('District_name')['Population'].sum().sort_values(ascending = False).head(20)
df10
```

```python
plt.figure(figsize=(10,6))
plt.xticks(rotation=90)
ax = sns.lineplot(x = df10.index, y = df10.values)
ax.ticklabel_format(style='plain', axis='y')
plt.ylabel('Population')
plt.title('District wise Total population')
plt.show()
```

# 3. Total Religious distribution¶

```python
df3 = df[['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].sum()
df3
```

```python
df3.plot(y = df3.values,kind = 'pie',autopct= '%1.1f%%')
plt.show()
```

# 4. Relious Population State wise

```python
df4 = df.groupby('State_name')[['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains']].sum()
df4
```

```python

sns.heatmap(df4, cmap='viridis' )
plt.show()
```

# 5. Education Level

```python
df5 = df[['Secondary_Education', 'Higher_Education', 'Graduate_Education']].sum()
```

```python
df5
```

```python
plt.xticks(rotation=90)
ax = sns.barplot(x = df5.values, y = df5.index)
ax.ticklabel_format(style='plain', axis='x')
plt.show()
```

# 6. Workers Type distribution

```python
df6 = df.groupby('State_name')[['Literate', 'Workers', 'Male_Workers', 'Female_Workers',
       'Cultivator_Workers', 'Agricultural_Workers', 'Household_Workers']].sum().sort_values(by = 'Literate',ascending = False)
```

```python
df6
```

```python
df7 = df.groupby('State_name')[['Male_Workers', 'Female_Workers', 'Agricultural_Workers', 'Household_Workers']].sum().sort_values(by = 'Male_Workers', ascending = False).head(6)
df7
```

```python
ax = df7.plot(kind = 'bar', stacked = True)
ax.ticklabel_format(style='plain', axis='y')
```

# 7. Age distribution

```python
df8 = df[['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50']].sum()
df8
```

```python
df8.plot(kind = 'pie', autopct = '%1.2f%%')
plt.show()
```

```python
df9 = df.groupby('State_name')[['Age_Group_0_29', 'Age_Group_30_49', 'Age_Group_50']].sum()
df9
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size
plt.hist([df9['Age_Group_0_29'], df9['Age_Group_30_49'], df9['Age_Group_50']], bins=10, alpha=0.7, label=['0-29', '30-49', '50+'])

# Adding labels and title
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Distribution of Population by Age Group')
plt.legend()

plt.show()
```
